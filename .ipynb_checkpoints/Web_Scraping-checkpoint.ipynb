# Details of Job Posting from a Single Page

```python
for jobs in soup.find_all(class_ = 'result'):
    
    try:
        job_title = jobs.a.text.strip()
    
    except:
        job_title = None
        
    print('Job Title:', job_title)
    
    
    try:
        company_name = jobs.span.text.strip()
        
    except:
        company_name = None
        
    print('Company Name:', company_name)
    
    
    
    try:
        location = jobs.find('span', class_ = 'location').text.strip()
        
    except:
        location = None
        
    print('Company Location',location)
    
    try:
        salary = jobs.find('div', class_ = 'salarySnippet').text.strip()
    
    except:
        salary = None
    
    print('Salary:', salary)
    
    try:
        summary = jobs.find('div', class_ = 'summary').text.strip()
    
    except:
        summary = None
        
    print('Summary', summary)
    
    print(112 * '-')
```

# Details of Job Posting from entire Website

```python
pages = [0, 10, 20, 30, 40, 50, 60, 70, 80, 100, 110]

for page in pages:
    source = requests.get('https://www.indeed.co.in/jobs?q=data+science&start={}'.format(page)).text
    soup = BeautifulSoup(source, 'lxml')
    
    for jobs in soup.find_all(class_ = 'result'):
    
        try:
            job_title = jobs.a.text.strip()

        except:
            job_title = None

        print('Job Title:', job_title)


        try:
            company_name = jobs.span.text.strip()

        except:
            company_name = None

        print('Company Name:', company_name)



        try:
            location = jobs.find('span', class_ = 'location').text.strip()

        except:
            location = None

        print('Company Location',location)

        try:
            salary = jobs.find('div', class_ = 'salarySnippet').text.strip()

        except:
            salary = None

        print('Salary:', salary)

        try:
            summary = jobs.find('div', class_ = 'summary').text.strip()

        except:
            summary = None

        print('Summary', summary)

        print(112 * '-')
        
        time.sleep(2)
        
```

# Saving the scraped Data to a Csv File

```python
pages = [0, 10, 20, 30, 40, 50, 60, 70, 80, 100, 110]

with open('/home/admin1/data_science.csv', 'a', encoding = 'utf-8', newline = '') as f:
    csv_print = csv.writer(f)
    
    file_is_empty = os.stat('/home/admin1/data_science.csv').st_size == 0
    
    if file_is_empty:
        csv_print.writerow(['Job Title', 'Company', 'Location', 'Salary', 'Summary'])
        
        
    for page in pages:
        source = requests.get('https://www.indeed.co.in/jobs?q=data+science&start={}'.format(page)).text
        soup = BeautifulSoup(source, 'lxml')

        for jobs in soup.find_all(class_ = 'result'):

            try:
                job_title = jobs.a.text.strip()

            except:
                job_title = None

            print('Job Title:', job_title)


            try:
                company_name = jobs.span.text.strip()

            except:
                company_name = None

            print('Company Name:', company_name)



            try:
                location = jobs.find('span', class_ = 'location').text.strip()

            except:
                location = None

            print('Company Location',location)

            try:
                salary = jobs.find('div', class_ = 'salarySnippet').text.strip()

            except:
                salary = None

            print('Salary:', salary)

            try:
                summary = jobs.find('div', class_ = 'summary').text.strip()

            except:
                summary = None

            print('Summary', summary)
            
            csv_print.writerow([job_title, company_name, location, salary, summary])
            
            print(112 * '-')

            time.sleep(2)

```
